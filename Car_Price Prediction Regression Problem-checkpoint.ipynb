## Converting Categorical Data to Numerical

# Concated The Selected Numerical & Categorical Features

```python
final_selcted_features=pd.concat([numeric_selct_data.drop(['price'],axis=1),selct_cat,data1['price']],axis=1)
final_selcted_features.head()
```

## Modelling 

```python
X=final_selcted_features.iloc[:,:-1]
y=final_selcted_features['price']

from sklearn.model_selection import train_test_split

X_train,X_test, y_train,y_test =train_test_split(X,y,test_size=0.20, random_state=101)
```

## To Check all the Statistical Relation Between Independent and Dependent Features

```python
import statsmodels.api as sm

X = sm.add_constant(X)
model= sm.OLS(y, X).fit()
```

```python
model.summary()
```

## Linear Regression

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X,y)
pred=lr.predict(X_test)

# The coefficients
print ('Coefficients: ', lr.coef_)
print ('Intercept: ',lr.intercept_)


print("Mse:",mean_squared_error(y_test,pred))
print('Mean Sq Log Error:- ',np.log(mean_squared_error(y_test,pred)))
rms = np.sqrt(mean_squared_error(y_test,pred))
print("Rmse:",rms)
print("Mape:",np.mean(np.abs(y_test-pred)/y_test)*100)
print("R-square:",np.round(r2_score(y_test,pred),2))
```

## Using Ensembling Technique( Random Forest  )

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=200,random_state=101)
rf.fit(X,y)
pred = rf.predict(X_test)

# # The coefficients (andomForestRegressor' object has no attribute 'coef_')
# print ('Coefficients: ', rf.coef_)
# print ('Intercept: ',rf.intercept_)

print("Mse:",mean_squared_error(y_test,pred))
rms =np.sqrt(mean_squared_error(y_test,pred))
print("Rmse:",rms)
# print("Mape:",mean_absolute_percentage_error(y_test,pred))
print("Mape:",np.mean(np.abs(y_test-pred)/y_test)*100)
print('Mean Sq Log Error:- ',np.log(mean_squared_error(y_test,pred)))
print("R-square:",np.round(r2_score(y_test,pred),2))
```

## Using Boosting Technique (XG Boost)

```python
import xgboost as Xgb
# from sklearn.model_selection import GridSearchCV 
xgb = Xgb.XGBRegressor(learning_rate=0.3)

# xgb_grid = GridSearchCV(xgb,parameters,cv = 2)
# xgb_grid.fit(X,y)
# pred = xgb_grid.predict(X_test)
xgb.fit(X,y)
pred=xgb.predict(X_test)

# print(xgb_grid.best_score_)
# print(xgb_grid.best_ntree_limit)

print("Mse:",mean_squared_error(y_test,pred))
rms = np.sqrt(mean_squared_error(y_test,pred))
print("Rmse:",rms)
print("Mape:",np.mean(np.abs(y_test-pred)/y_test)*100)
print('Mean Sq Log Error:- ',np.log(mean_squared_error(y_test,pred)))
print("R-square:",np.round(r2_score(y_test,pred),2))
```
